```python
import csv
```

```python
with open('lab.csv',"r",encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file)
    l = list(csv_reader)
```

```python
print("First 20 lines : ")
s = 0
for x in l[:20]:
    s+=1
    print(s , (" ".join(x)).strip())
print()
print("Last 20 lines : ")
s = 0
print()
for i in l[-20:]:
    s+=1
    print(s , (" ".join(i)).strip())
```

# QUESTION 2

```python
ls = []
for x in l:
    ls.append((" ".join(x)).strip().split())
```

```python
for x in ls:
    while "question" in x:
        ind = x.index("question")
        print(ls.index(x),x)
        x[ind] = "answer"
        print(x)
```
