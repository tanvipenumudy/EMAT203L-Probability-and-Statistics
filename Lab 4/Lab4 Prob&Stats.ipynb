```python
#Question4
pt=0.5
ph=0.5
total=100
def choose(n,k):
    if(0<=k<=n):
        ntok = 1
        ktok = 1
        for t in range(1,min(k, n-k)+1):
            ntok*=n
            ktok*=t
            n-=1
        return ntok//ktok
    else:
        return 0
p30=choose(total,30)*(pt**total)    
print("Probability of getting 30 tails exactly",p30)
net = 0
for i in range(36):
    net = net + choose(total,i)*(pt**total)
pg35 = 1 - net
print("Probability of getting more than 35 tails",pg35)
p60 = 0
for i in range(61,70):
    p60= p60 + choose(total,i)*(pt**total)
print("Probability of getting tails in between 60 and 70",p60)
```
